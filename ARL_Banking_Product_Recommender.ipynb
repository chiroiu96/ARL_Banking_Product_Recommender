```
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

#Read Ratings DataFrame

```
df1 = pd.read_csv('https://raw.githubusercontent.com/chiroiu96/ARL_Banking_Product_Recommender/main/train_ARL_1.csv', sep=',')
df2 = pd.read_csv('https://raw.githubusercontent.com/chiroiu96/ARL_Banking_Product_Recommender/main/train_ARL_2.csv', sep=',')
df = df1.append(df2, ignore_index=True)
df.head(10)
```

#Binary Encoding of dataframe

```
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
df = df.applymap(encode_units)
```

#Generation of most frequent items

```
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)
```

#Generation of association rules

```
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules.head()
```

#Choosing the rules with high impact
We are interested in rules:


*   Have a support as hight as possible (value depende on number of items)
*   Have a confidence close to 1
*   Have a lift higher then 1 (positive correlation)



```
rules[ (rules['lift'] >= 6) &
      (rules['confidence'] >= 0.8) ]
```
